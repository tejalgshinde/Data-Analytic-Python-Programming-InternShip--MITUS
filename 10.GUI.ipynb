```python
import tkinter as tk
from tkinter import BooleanVar
from tkinter.ttk import *
from tkinter import scrolledtext
from tkinter import messagebox

top = tk.Tk()
top.title("Vishay Kholll")
top.geometry('500x400')
e1 = tk.Entry(top).grid(column = 0, row = 1)
b1 = tk.Button(top, text = "Click here", bg = "White").grid(column = 0, row = 2)
t1 = tk.Text(top).grid(column = 1, row = 3)
l1 = tk.Label(top, text = "Welcome", font = ('Ariel Bold', 50)).grid(column = 2, row = 2)

#combo box
combo = Combobox(top)
combo['values'] = [1,2,3,4, 'next']
combo.current(2)
combo.grid(column = 0, row = 0)

#check button
ch = BooleanVar()
ch.set(True)
ch = Checkbutton(top, text = 'Done', var = ch, width = 30).grid(column = 3, row = 1)

#scrolled text
st = scrolledtext.ScrolledText(top, width = 40, height = 10).grid(column = 3, row = 3)

top.mainloop()
```

```python
# square of a number

top = tk.Tk()

top.title('Square')
top.geometry('500x400')

a = tk.IntVar()
def square():
    num = a.get()
    sq = num * num
    result = tk.Label(top, text = 'Square is %i' %sq).place(x = 130, y = 190)
    messagebox.showinfo("result", "Square is %i" %sq)

l = tk.Label(top, text = 'INPUT').place(x = 30, y = 100)
e = tk.Entry(top, textvariable = a).place(x = 100, y = 100)
b = tk.Button(top, text = 'Calculate', font = ('Ariel Bold', 15),command = square).place(x = 120,y = 150)

top.mainloop()
```

```python

import tkinter as tk
from tkinter import BooleanVar
from tkinter.ttk import *
from tkinter import scrolledtext
from tkinter import messagebox

bt = tk.Tk()
bt.title("Operations")
bt.geometry('500x400')

a = tk.IntVar()
b = tk.IntVar()
def addition():
    a1 = a.get()
    b2 = b.get()
    add = a1 + b2
    result = tk.Label(bt, text = 'Addition is %i' %add).place(x = 130, y = 190)
    messagebox.showinfo("result", "Additon is %i" %add)

def substraction():
    a1 = a.get()
    b2 = b.get()
    subt = a1 - b2
    result = tk.Label(bt, text = 'Substraction is %i' %subt).place(x = 130, y = 190)
    messagebox.showinfo("result", "Substraction is %i" %subt)

def multiplication():
    a1 = a.get()
    b2 = b.get()
    mul = a1 * b2
    result = tk.Label(bt, text = 'Multiplication is %i' %mul).place(x = 130, y = 190)
    messagebox.showinfo("result", "Multiplication is %i" %mul)

def division():
    a1 = a.get()
    b2 = b.get()
    divo = a1 / b2
    result = tk.Label(bt, text = 'Division is %i' %divo).place(x = 130, y = 190)
    messagebox.showinfo("result", "Division is %i" %divo)



l = tk.Label(bt, text = 'INPUT').place(x = 30, y = 100)
e = tk.Entry(bt, textvariable = a).place(x = 100, y = 100)
l1 = tk.Label(bt, text = 'INPUT').place(x = 30, y = 130)
e1 = tk.Entry(bt, textvariable = b).place(x = 100, y = 130)
b5 = tk.Button(bt, text = 'Addition', font = ('Ariel Bold', 15),command = addition).place(x = 120,y = 150)
b6 = tk.Button(bt, text = 'Substraction', font = ('Ariel Bold', 15),command = substraction).place(x = 120,y = 210)
b7 = tk.Button(bt, text = 'Multiplication', font = ('Ariel Bold', 15),command = multiplication).place(x = 120,y = 270)
b10 = tk.Button(bt, text = 'Division', font = ('Ariel Bold', 15),command = division).place(x = 120,y = 330)

bt.mainloop()
```
