```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.array([10,20,30,60,50,80,90,40,70,100])
y = np.array([20,41,52,95,84,75,45,62,52,53])
z = np.array([45,85,63,52,41,85,74,96,85,15])
```

### Line Plot

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x, y, marker = 'o', mfc = 'red', color = 'black', ls = '--', ms = 5);
plt.grid()
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x, y, marker = 'o', mfc = 'red', color = 'black', ls = '--', ms = 5,label = '2023');
plt.plot(x, z, marker = '<', mfc = 'red', color = 'blue', ls = '--',ms = 7,label = '2024');
plt.grid()
plt.legend();
```

```python
# subplot

plt.figure(figsize=(16,10))

plt.subplot(2,2,1)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x, y, marker = 'o', mfc = 'red', color = 'black', ls = '--', ms = 5);
plt.grid()

plt.subplot(2,2,2)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x, y, marker = 'o', mfc = 'red', color = 'black', ls = '--', ms = 5,label = '2023');
plt.plot(x, z, marker = '<', mfc = 'red', color = 'blue', ls = '--',ms = 7,label = '2024');
plt.grid()
plt.legend();






```

### Box Plot

```python
marks = [52,62,85,96,64,62,65,47,55,91]
plt.boxplot(marks);
```

```python
q3 = np.percentile(marks, 75)
q1 = np.percentile(marks, 25)
q3, q1
```

```python
iqr = q3 - q1
iqr
```

```python
ub = q3 + (1.5 * iqr)
lb = q1 - (1.5 * iqr)
```
